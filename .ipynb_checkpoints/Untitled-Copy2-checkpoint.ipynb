## 2.2 Pembersihan Teks

```python
#import regular expression untuk melakukan pembersihan data
import re
```

```python
#membuat fungsi untuk pembersihan teks yang dapat digunakan untuk inputan user dan pemebrsihan data training
def pembersih_text(text):
    #membuat semua teks jadi huruf kecil
    text = text.lower()
    
    #menghapus special karaketer dari teks
    text_bersih = re.sub(r'[^\w\s]', '', text)
    
    #mengpus spasi ekstra
    text_bersih = " ".join(text_bersih.split())
    
    #pengembalian nilai kedalam variabel teks_bersih
    return text_bersih
```

```python
#melakukan pemebersihan pada dataset training kolom teks
data_train["text"] = data_train["text"].apply(pembersih_text)
```

```python
#melakukan pemebersihan pada dataset training kolom teks
data_valid["text"] = data_valid["text"].apply(pembersih_text)
```

```python
from tensorflow.keras.layers import Input, Embedding, Conv1D, MaxPooling1D, GlobalMaxPooling1D, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Mengambil teks dan label dari DataFrame training, validasi, dan testing
train_texts = data_train['text'].tolist()
train_labels = data_train['label'].tolist()
valid_texts = data_valid['text'].tolist()
valid_labels = data_valid['label'].tolist()
test_texts = data_testing['text'].tolist()
test_labels = data_testing['label'].tolist()

# Tokenisasi teks
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train_texts)

# Mengonversi teks ke dalam urutan token
train_sequences = tokenizer.texts_to_sequences(train_texts)
valid_sequences = tokenizer.texts_to_sequences(valid_texts)
test_sequences = tokenizer.texts_to_sequences(test_texts)

# Padding urutan token
maxlen = 100  # Misalkan panjang maksimum token
train_padded = pad_sequences(train_sequences, maxlen=maxlen, padding='post')
valid_padded = pad_sequences(valid_sequences, maxlen=maxlen, padding='post')
test_padded = pad_sequences(test_sequences, maxlen=maxlen, padding='post')

# Mengonversi label pelatihan, validasi, dan testing menjadi format numerik
label_encoder = LabelEncoder()
train_labels_encoded = label_encoder.fit_transform(train_labels)
valid_labels_encoded = label_encoder.transform(valid_labels)
test_labels_encoded = label_encoder.transform(test_labels)

# Membangun model CNN
V = len(tokenizer.word_index)  # Jumlah kata dalam kosakata
D = 20  # Dimensi embedding
T = maxlen  # Panjang maksimum urutan token

# Definisi input layer
i = Input(shape=(T,))

# Definisi layer embedding
x = Embedding(input_dim=V+1, output_dim=D)(i)

# Convolutional layer pertama
x = Conv1D(32, 3, activation='relu')(x)
x = MaxPooling1D(3)(x)

# Convolutional layer kedua
x = Conv1D(64, 3, activation='relu')(x)
x = MaxPooling1D(3)(x)

# Convolutional layer ketiga
x = Conv1D(128, 3, activation='relu')(x)

# Global Max Pooling layer
x = GlobalMaxPooling1D()(x)

# Dense layer dengan softmax activation
x = Dense(3, activation='softmax')(x)  # Output memiliki 3 kelas (sesuaikan dengan jumlah kelas Anda)

# Membangun model
model = Model(inputs=i, outputs=x)

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Melatih model
history = model.fit(train_padded, train_labels_encoded, epochs=100, validation_data=(valid_padded, valid_labels_encoded))

# Evaluasi model pada data testing
test_loss, test_accuracy = model.evaluate(test_padded, test_labels_encoded)
print("Test Accuracy:", test_accuracy)

# Evaluasi model pada data validasi
valid_predictions_probs = model.predict(valid_padded)
valid_predictions = valid_predictions_probs.argmax(axis=-1)

# Menghasilkan matriks kebingungan
conf_matrix = confusion_matrix(valid_labels_encoded, valid_predictions)

# Visualisasi matriks kebingungan
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix (Validation)')
plt.show()

# Tampilkan laporan klasifikasi
class_report = classification_report(valid_labels_encoded, valid_predictions, target_names=label_encoder.classes_)
print("Classification Report (Validation):")
print(class_report)

# Visualisasi akurasi dan loss
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0, 1])
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')
plt.show()

plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label = 'val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
def preprocess_text(text):
    # Tokenisasi teks
    sequence = tokenizer.texts_to_sequences([text])
    # Padding urutan token
    padded_sequence = pad_sequences(sequence, maxlen=maxlen, padding='post')
    return padded_sequence

def classify_text(text):
    # Prediksi kelas
    text_sequence = preprocess_text(text)
    prediction_prob = model.predict(text_sequence)[0]
    predicted_class_index = prediction_prob.argmax()
    predicted_class = label_encoder.classes_[predicted_class_index]
    return predicted_class

# Pengujian model dengan input pengguna
# user_input = input("Masukkan teks untuk diklasifikasikan: ")
predicted_class = classify_text(text='kita akan semakin cerdas jika kita terus menerus melatih otak kita')
print("Hasil klasifikasi:", predicted_class)
```
