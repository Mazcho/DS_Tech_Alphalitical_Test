# Load DataSet

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# library baca dataset
import pandas as pd

#memberikan nama kolom ( karena pada dataset tidak memiliki nama kolom)
nama_kolom = ["teks","label"]

train_df = pd.read_csv('train_preprocess.tsv', sep="\t", header=None, names=nama_kolom)
test_df = pd.read_csv('test_preprocess.tsv', sep="\t", header=None, names=nama_kolom)
valid_df = pd.read_csv('valid_preprocess.tsv', sep="\t", header=None, names=nama_kolom)
```

# Data Cleansing PreProcesing

Pada tahap Data Cleansing dan PreProcesing data, saya mempunyai beberapa langkah sebagai berikut.
- 1. **Pengecekan Missing Value** : Melakukan pengecekan missing value terhadap 3 data set yang telah dberikan, yaitu data training, testing, dan validasi

- 2. **Pembersihan Teks** : Pada tahap ini, saya melakukan pembersihan teks dengan cara membuat semua teks menjadi huruf kecil, penghapusan angka dan penghapusa tanda baca/ akrakter special dan spasi.

- 3. **Tokenisasi** : Setelah melakukan pembersihan teks, dilakukan tokenisasi yang bertujuan mengubah teks jadi urutan angka menggunakan TOKENIZER lalu mengubah teks ke sequence
- 4. **Padding** : Setelah tokenisasi dilakukan penyelarasan panjang urutan teks
- 5. **Label Encoder** : Melakukan labeling pada kelas label. Mengubahnya menjadi angka. 

_**Note**_ : Saya tidak memasukkan Stopword dari NLTK dan Steming dari Sastrawi Indonesia dikarenakan merusak makna dalam teks. Saya telah mencoba beberapa kali di eksperimen pengerjaan ini.

```python
#[engecekan missing Value
print("Nilai missing value pada data_train :\n", train_df.isna().sum())
print("Nilai missing value pada data_test :\n", test_df.isna().sum())
print("Nilai missing value pada data_valid :\n", valid_df.isna().sum())
```

```python
#import librari dulu
import re #ini regular express buat bersih bersih
import string #ini string buat penghapusan spasi


#pembersihan Data
def bersihin_teks(text):
    text = text.lower()  #mengubah teks jadi kecil smua
    text = re.sub(r'\d+', '', text)#hapus angka
    text = text.translate(str.maketrans('', '', string.punctuation))#buat hapus tanda baca
    text = text.strip()#hapus spasis di awal dan ahkir
    return text

#menerapkan fungsi kepada masing masing teks
train_df['teks'] = train_df['teks'].apply(bersihin_teks)
valid_df['teks'] = valid_df['teks'].apply(bersihin_teks)
test_df['teks'] = test_df['teks'].apply(bersihin_teks)
```

```python
#ini proses tokenisasi dan padding

#import librari nya dulu
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer #buat tokenisasi
from tensorflow.keras.preprocessing.sequence import pad_sequences#buat padding

#melakukan tokenisasi 
tokenizer = Tokenizer(num_words=5000, oov_token="<OOV>")
tokenizer.fit_on_texts(train_df['teks'])
word_index = tokenizer.word_index

#mengubah teks menjadi sequence
train_sequences = tokenizer.texts_to_sequences(train_df['teks'])
valid_sequences = tokenizer.texts_to_sequences(valid_df['teks'])
test_sequences = tokenizer.texts_to_sequences(test_df['teks'])

#ini untuk padding sequence
max_length = 200
train_padded = pad_sequences(train_sequences, padding='post', maxlen=max_length)
valid_padded = pad_sequences(valid_sequences, padding='post', maxlen=max_length)
test_padded = pad_sequences(test_sequences, padding='post', maxlen=max_length)
```

```python
# #menyimpan tokenizer yang sudah di proses
# import pickle

# with open('tokenizer_baru.pickle', 'wb') as handle:
#     pickle.dump(tokenizer, handle, protocol=pickle.HIGHEST_PROTOCOL)

# print("Tokenizer telah disimpan setelah proses tokenisasi selesai.")
```

```python
#melakukan label encode pada target (label)
#import label encoder
from sklearn.preprocessing import LabelEncoder

#load label encodernya
label_encoder = LabelEncoder()

#penerapannya ke dalam data
y_train = label_encoder.fit_transform(train_df['label'])
y_valid = label_encoder.transform(valid_df['label'])
y_test = label_encoder.transform(test_df['label'])
```

# Membangun Model

Pada pengerjaan tugas ini, saya pakai model CNN, RNN, dan LSTM

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, Dense, LSTM, SimpleRNN, Dropout
```

```python
# Model CNN
model_cnn = Sequential([
    Embedding(input_dim=5000, output_dim=64, input_length=max_length),
    Conv1D(128, 5, activation='relu'),
    GlobalMaxPooling1D(),
    Dense(10, activation='relu'),
    Dense(3, activation='softmax')
])

model_cnn.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model_cnn.summary()
```

```python
# Model RNN
model_rnn = Sequential([
    Embedding(input_dim=5000, output_dim=64, input_length=max_length),
    SimpleRNN(128, return_sequences=False),
    Dense(10, activation='relu'),
    Dense(3, activation='softmax')
])

model_rnn.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model_rnn.summary()
```

```python
#model LSTM
model_lstm = Sequential([
    Embedding(input_dim=5000, output_dim=64, input_length=max_length),
    LSTM(128, return_sequences=False),
    Dense(10, activation='relu'),
    Dense(3, activation='softmax')
])

model_lstm.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model_lstm.summary()
```

# Melatih Model CNN dan evaluasi CNN

```python
#import buat metrix evaluasi
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
#melatihan Model CNN
history_cnn = model_cnn.fit(train_padded, y_train, epochs=10, validation_data=(valid_padded, y_valid), verbose=2)
```

```python
# Evaluasi Model CNN
y_pred_cnn = np.argmax(model_cnn.predict(test_padded), axis=1)

print("CNN Accuracy (Testing):", accuracy_score(y_test, y_pred_cnn))
print("Classification Report CNN (Testing):\n", classification_report(y_test, y_pred_cnn, target_names=label_encoder.classes_))

# Menyimpan nilai akurasi CNN
akurasi_CNN = accuracy_score(y_test, y_pred_cnn)

# Visualisasi Confusion Matrix menggunakan seaborn
confusion_matrix_hasil = confusion_matrix(y_test, y_pred_cnn)
plt.figure(figsize=(10, 7))
sns.heatmap(data=confusion_matrix_hasil, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix - CNN')
plt.show()

# Plot sejarah akurasi
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
plt.plot(history_cnn.history['accuracy'], label='Train Accuracy')
plt.plot(history_cnn.history['val_accuracy'], label='Val Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Model Accuracy')

# Plotting Loss
plt.subplot(1, 2, 2)
plt.plot(history_cnn.history['loss'], label='Train Loss')
plt.plot(history_cnn.history['val_loss'], label='Val Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.title('Model Loss')

plt.show()
```
