## 2.2 Pembersihan Teks

```python
#import regular expression untuk melakukan pembersihan data
import re
```

```python
#membuat fungsi untuk pembersihan teks yang dapat digunakan untuk inputan user dan pemebrsihan data training
def pembersih_text(text):
    #membuat semua teks jadi huruf kecil
    text = text.lower()
    
    #menghapus special karaketer dari teks
    text_bersih = re.sub(r'[^\w\s]', '', text)
    
    #mengpus spasi ekstra
    text_bersih = " ".join(text_bersih.split())
    
    #pengembalian nilai kedalam variabel teks_bersih
    return text_bersih
```

```python
#melakukan pemebersihan pada dataset training kolom teks
data_train["text"] = data_train["text"].apply(pembersih_text)
```

```python
#melakukan pemebersihan pada dataset training kolom teks
data_valid["text"] = data_valid["text"].apply(pembersih_text)
```

## 2.4 Melakukan Stemming pada data bahasa indonesia menggunakan library Sastrawi

Melakukan setting dulu untuk stemming bahasa indonesia pakai sastrawi

```python
#import library
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Inisialisasi objek stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def steming_text(text):
    return stemmer.stem(text)
```

```python
#melakukan steming teks indonesia
data_train["text"] = data_train["text"].apply(steming_text)
```

```python
#melakukan steming teks indonesia
data_valid["text"] = data_valid["text"].apply(steming_text)
```

```python
from tensorflow.keras.layers import Input, Embedding, Conv1D, MaxPooling1D, GlobalMaxPooling1D, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Misalkan df_training adalah DataFrame untuk dataset pelatihan dan df_validasi adalah DataFrame untuk dataset validasi

# Mengambil teks dan label dari DataFrame training dan validasi
train_texts = data_train['text'].tolist()
train_labels = data_train['label'].tolist()
valid_texts = data_valid['text'].tolist()
valid_labels = data_valid['label'].tolist()

# Tokenisasi teks
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train_texts)

# Mengonversi teks ke dalam urutan token
train_sequences = tokenizer.texts_to_sequences(train_texts)
valid_sequences = tokenizer.texts_to_sequences(valid_texts)

# Padding urutan token
maxlen = 100  # Misalkan panjang maksimum token
train_padded = pad_sequences(train_sequences, maxlen=maxlen, padding='post')
valid_padded = pad_sequences(valid_sequences, maxlen=maxlen, padding='post')

# Mengonversi label pelatihan dan validasi menjadi format numerik
label_encoder = LabelEncoder()
train_labels_encoded = label_encoder.fit_transform(train_labels)
valid_labels_encoded = label_encoder.transform(valid_labels)

# Membangun model CNN
V = len(tokenizer.word_index)  # Jumlah kata dalam kosakata
D = 20  # Dimensi embedding
T = maxlen  # Panjang maksimum urutan token

# Definisi input layer
i = Input(shape=(T,))

# Definisi layer embedding
x = Embedding(input_dim=V+1, output_dim=D)(i)

# Convolutional layer pertama
x = Conv1D(32, 3, activation='relu')(x)
x = MaxPooling1D(3)(x)

# Convolutional layer kedua
x = Conv1D(64, 3, activation='relu')(x)
x = MaxPooling1D(3)(x)

# Convolutional layer ketiga
x = Conv1D(128, 3, activation='relu')(x)

# Global Max Pooling layer
x = GlobalMaxPooling1D()(x)

# Dense layer dengan softmax activation
x = Dense(5, activation='softmax')(x)  # Misalkan output memiliki 5 kelas

# Membangun model
model = Model(inputs=i, outputs=x)

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Melatih model
history = model.fit(train_padded, train_labels_encoded, epochs=100, validation_data=(valid_padded, valid_labels_encoded))

# Evaluasi model
valid_predictions_probs = model.predict(valid_padded)
valid_predictions = valid_predictions_probs.argmax(axis=-1)

# Menghasilkan matriks kebingungan
conf_matrix = confusion_matrix(valid_labels_encoded, valid_predictions)

# Visualisasi matriks kebingungan
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()

# Tampilkan laporan klasifikasi
class_report = classification_report(valid_labels_encoded, valid_predictions, target_names=label_encoder.classes_)
print("Classification Report:")
print(class_report)

# Visualisasi akurasi dan loss
import matplotlib.pyplot as plt

# Visualisasi akurasi
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0, 1])
plt.legend(loc='lower right')
plt.show()

# Visualisasi loss
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label = 'val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.show()
```

```python
def preprocess_text(text):
    # Tokenisasi teks
    sequence = tokenizer.texts_to_sequences([text])
    # Padding urutan token
    padded_sequence = pad_sequences(sequence, maxlen=maxlen, padding='post')
    return padded_sequence

def classify_text(text):
    # Prediksi kelas
    text_sequence = preprocess_text(text)
    prediction_prob = model.predict(text_sequence)[0]
    predicted_class_index = prediction_prob.argmax()
    predicted_class = label_encoder.classes_[predicted_class_index]
    return predicted_class

# Pengujian model dengan input pengguna
user_input = input("Masukkan teks untuk diklasifikasikan: ")
predicted_class = classify_text(user_input)
print("Hasil klasifikasi:", predicted_class)
```

```python
import re
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Inisialisasi objek stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def pembersih_text(text):
    #membuat semua teks jadi huruf kecil
    text = text.lower()
    
    #menghapus special karaketer dari teks
    text_bersih = re.sub(r'[^\w\s]', '', text)
    
    #mengapus spasi ekstra
    text_bersih = " ".join(text_bersih.split())
    
    #pengembalian nilai kedalam variabel teks_bersih
    return text_bersih

def steming_text(text):
    return stemmer.stem(text)

def preprocess_user_input(text, tokenizer, maxlen):
    text_bersih = pembersih_text(text)
    text_stem = steming_text(text_bersih)
    
    # Tokenisasi teks
    sequence = tokenizer.texts_to_sequences([text_stem])
    # Padding urutan token
    padded_sequence = pad_sequences(sequence, maxlen=maxlen, padding='post')
    
    return padded_sequence

# Pengujian model dengan input pengguna
user_input = input("Masukkan teks untuk diklasifikasikan: ")
user_input_preprocessed = preprocess_user_input(user_input, tokenizer, maxlen)
predicted_class_prob = model.predict(user_input_preprocessed)[0]
predicted_class_index = predicted_class_prob.argmax()
predicted_class = label_encoder.classes_[predicted_class_index]
print("Hasil klasifikasi:", predicted_class)
```

```python
from transformers import BertTokenizer, TFBertForSequenceClassification
import tensorflow as tf
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Load pre-trained BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Tokenisasi teks
train_tokenized = tokenizer(train_texts, padding=True, truncation=True, return_tensors="tf")
valid_tokenized = tokenizer(valid_texts, padding=True, truncation=True, return_tensors="tf")

# Mengonversi label pelatihan dan validasi menjadi format numerik
label_encoder = LabelEncoder()
train_labels_encoded = label_encoder.fit_transform(train_labels)
valid_labels_encoded = label_encoder.transform(valid_labels)

# Load pre-trained BERT model for sequence classification
model = TFBertForSequenceClassification.from_pretrained('bert-base-uncased')

# Compile model
optimizer = tf.keras.optimizers.Adam(learning_rate=3e-5)
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])

# Melatih model
history = model.fit(train_tokenized, train_labels_encoded, epochs=3, validation_data=(valid_tokenized, valid_labels_encoded))

# Evaluasi model
valid_predictions_probs = model.predict(valid_tokenized)
valid_predictions = valid_predictions_probs.logits.argmax(axis=-1)

# Menghasilkan matriks kebingungan
conf_matrix = confusion_matrix(valid_labels_encoded, valid_predictions)

# Visualisasi matriks kebingungan
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()

# Tampilkan laporan klasifikasi
class_report = classification_report(valid_labels_encoded, valid_predictions, target_names=label_encoder.classes_)
print("Classification Report:")
print(class_report)

# Visualisasi akurasi dan loss
# Plot accuracy
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Plot loss
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import Embedding, Conv1D, MaxPooling1D, LSTM, Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Mengambil teks dan label dari DataFrame training dan validasi
train_texts = data_train['text'].tolist()
train_labels = data_train['label'].tolist()
valid_texts = data_valid['text'].tolist()
valid_labels = data_valid['label'].tolist()

# Tokenisasi teks
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train_texts)

# Mengonversi teks ke dalam urutan token
train_sequences = tokenizer.texts_to_sequences(train_texts)
valid_sequences = tokenizer.texts_to_sequences(valid_texts)

# Padding urutan token
maxlen = 100  # Misalkan panjang maksimum token
train_padded = pad_sequences(train_sequences, maxlen=maxlen, padding='post')
valid_padded = pad_sequences(valid_sequences, maxlen=maxlen, padding='post')

# Mengonversi label pelatihan dan validasi menjadi format numerik
label_encoder = LabelEncoder()
train_labels_encoded = label_encoder.fit_transform(train_labels)
valid_labels_encoded = label_encoder.transform(valid_labels)

# One-hot encoding untuk label
train_labels_encoded = tf.keras.utils.to_categorical(train_labels_encoded, num_classes=len(label_encoder.classes_))
valid_labels_encoded = tf.keras.utils.to_categorical(valid_labels_encoded, num_classes=len(label_encoder.classes_))

# Mengatur seed untuk reproduktibilitas
seed = 42
np.random.seed(seed)
tf.random.set_seed(seed)

# Mendefinisikan arsitektur model
def get_model(max_features, embed_dim):
    model = Sequential()
    model.add(Embedding(max_features, embed_dim, input_length=maxlen))
    model.add(Conv1D(filters=32, kernel_size=3, padding='same', activation='relu'))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Conv1D(filters=32, kernel_size=3, padding='same', activation='relu'))
    model.add(MaxPooling1D(pool_size=2))    
    model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))
    model.add(Dense(len(label_encoder.classes_), activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    print(model.summary())
    return model

# Menentukan parameter model
max_features = len(tokenizer.word_index) + 1
embed_dim = 100

# Membuat model
model = get_model(max_features, embed_dim)

# Melatih model
history = model.fit(train_padded, train_labels_encoded, epochs=10, validation_data=(valid_padded, valid_labels_encoded))

# Evaluasi model
valid_predictions_probs = model.predict(valid_padded)
valid_predictions = valid_predictions_probs.argmax(axis=-1)

# Menghasilkan matriks kebingungan
conf_matrix = confusion_matrix(valid_labels_encoded.argmax(axis=-1), valid_predictions)

# Visualisasi matriks kebingungan
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()

# Tampilkan laporan klasifikasi
class_report = classification_report(valid_labels_encoded.argmax(axis=-1), valid_predictions, target_names=label_encoder.classes_)
print("Classification Report:")
print(class_report)

# Visualisasi akurasi dan loss
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0, 1])
plt.legend(loc='lower right')
plt.show()

plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.show()

# Fungsi untuk mengklasifikasikan input teks pengguna
def preprocess_user_input(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    text = " ".join(text.split())
    return text

def classify_text(model, tokenizer, label_encoder, text):
    seq = tokenizer.texts_to_sequences([text])
    padded = pad_sequences(seq, maxlen=maxlen, padding='post')
    pred = model.predict(padded).argmax(axis=-1)
    return label_encoder.inverse_transform(pred)[0]

user_input = input("Masukkan teks untuk diklasifikasikan: ")
user_input_preprocessed = preprocess_user_input(user_input)
predicted_class = classify_text(model, tokenizer, label_encoder, user_input_preprocessed)
print("Hasil klasifikasi:", predicted_class)
```
