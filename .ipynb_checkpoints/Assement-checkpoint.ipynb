# 2. Data Cleansing dan Preprocesing

## 2.1 WorkFLow Cleansing dan PreProcessing

Sebelum Data diproses kedalam pemodelan, dilakukan cleansing dan Preprocesing. Hal ini dilakukan untuk mempersiapkan data agar mudah untuk diproses oleh model. Berikut alur Cleansing dan Preprocesing ebserta penjelasannya


- **1. Cleansing** : Pada tahap Data Cleansing, dilakukan pembersihan data seperti missing value, data yang duplikat, karakter karakter spesial seperti tanda baca, emoticon dan lainnya

- **Case Folding** : Pada tahap Case Folding, dilakukan pengubahan semua karakter dalam dataset menjadi huruf kecil semua. Jadi jika pada dataset "HALo NAMA saya niCHO" ----> "halo nama saya nicho". Serta melakukan penghapusan extra spasi

- **Stemming** : Stemming merupakan proses untuk mengubah kata yang memiliki imbuhan menjadi kata akar atau kata dasarnya.

- **Tokenisasi** : Tokenisasi adalah proses untuk memotong setiap kata pada data set menjadi kata yang independen

## 2.2 Kode Cleansing dan Pre-Processing

### 2.2.1 Cleansing Data

```python
#mengecek nilai missing value pada dataset train,test,valid
dataset = [data_train,data_testing,data_valid]

for i in dataset:
    if i.isna().sum().sum() == 0:
        print("Tidak ada nilai missing value")
        print(i.isna().sum())
        print("=============================")
    else:
        print("Ada missing value pada : ")
        print(i.isna().sum())
```

```python
#mengecek nilai duplikasi pada data train, testing, validasi
dataset = [data_train,data_testing,data_valid]

for i in dataset:
    if i.duplicated().sum() == 0:
        print("Tidak ada nilai duplikasi")
        print(i.duplicated().sum())
        print("=============================")
    else:
        print("Ada duplikasi : ")
        print(i.duplicated().sum())
        print("=================")
```

```python
#import regular expression untuk melakukan pembersihan pada kalimat
import re

#membuat fungsi untuk pembersihan teks yang dapat digunakan untuk inputan user dan pemebrsihan pada data 
def pembersih_text(text):
    
    #menghapus special karaketer dari teks
    text_bersih = re.sub(r'[^\w\s]', '', text)
    
    #pengembalian nilai kedalam variabel teks_bersih
    return text_bersih
```

```python
#melakukan pemebersihan pada dataset training kolom teks
data_train["text"] = data_train["text"].apply(pembersih_text)
```

### 2.2.2 Pre-Processing

```python
#Kode untuk case folding
def case_folding(text):
    #membuat semua teks jadi huruf kecil
    text_kecil = text.lower()
    
    #mengpus spasi ekstra
    text_bersih = " ".join(text_kecil.split())
    
    #pengembalian nilai kedalam variabel teks_bersih
    return text_bersih
```

```python
#melakukan pemebersihan pada dataset training kolom teks
data_train["text"] = data_train["text"].apply(case_folding)
```

```python
#melakukan pemebersihan pada dataset testing kolom teks
data_testing["text"] = data_testing["text"].apply(case_folding)
```

```python
#melakukan pemebersihan pada dataset valid kolom teks
data_valid["text"] = data_valid["text"].apply(case_folding)
```

Output dari DataRawCleasning ---> DataDataCleansingCaseFolding

```python
#kode untuk stemming pakai sastrawi

#import library
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Inisialisasi objek stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def steming_text(text):
    return stemmer.stem(text)
```

```python
#melakukan steming teks indonesia data training
data_train["text"] = data_train["text"].apply(steming_text)
```

```python
#melakukan steming teks indonesia data testing
data_testing["text"] = data_testing["text"].apply(steming_text)
```

```python
#melakukan steming teks indonesia data validasi
data_valid["text"] = data_valid["text"].apply(steming_text)
```
